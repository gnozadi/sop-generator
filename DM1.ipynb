```python
def full_proframme_to_semester(duration, price):
  # duration is in days
  month = duration / 12
  return price / 3
```

```python
def convert_to_semester(price_str, price, duration):
  if price_str == 'Tuition (Year)':
    return price / 2
  elif price_str == 'Tuition (Credit)':
    return price*15
  elif price_str == 'Tuition (Full programme)':
    return full_proframme_to_semester(duration, price)
  elif price_str == 'Tuition (Module)' or price_str == 'Tuition (Month)':
    return price * 4
  elif price_str == 'Tuition (Trimester)':
    return price * 3 / 2
  elif price_str == 'Tuition (Quarter)':
    return price * 2
  else:
    return price
```

```python
for i in range(program_count):
  if not pd.isna(result_df['tuition_price_specification'][i]):
    if not pd.isna(result_df['tution_1_money'][i]):
      result_df['tution_1_money'][i] = convert_to_semester(result_df['tuition_price_specification'][i],
                                                      result_df['tution_1_money'][i], 
                                                      result_df['duration'][i]
                                                        )
      result_df['tuition_price_specification'][i] = 'Tuition (Semester)'
```

```python
def get_gpa(s):
  gpa = 0
  if 'GPA' in s:
    if 'GPA of' in s:  
      gpa_index = s.split('GPA of')[1]
      if gpa_index.split()[0].isdigit():
        gpa = gpa_index.split()[0]
      elif 'better on a' in gpa_index:
        if len(gpa_index.split('better on a'))>1:
           gpa = 4.0
  return gpa
```

```python
def get_lang(i, x):
  lan = result_df['language'][i]
  
  if lan in x:
    return f'proficient in {lan}'
```

```python
def get_keywords(i, li_list):
  for i, x in enumerate(li_list):
    keyword_list = []
    x = x.text
    gpa = float(get_gpa(x))
    if gpa > 0:
      keyword_list.append(f'got gpa of {gpa}')
    if 'proficiency' in x:
      keyword_list.append(get_lang(i, x))
    if 'GRE' in x:
      keyword_list.append("took GRE exam")
  return keyword_list
```
